2. Create item profiles

```python
# read items file:

i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

items = pd.read_csv('./DATA/ml-100k/u.item', sep='|', names=i_cols,
 encoding='latin-1')
```

```python
# create feature matrix using TF-IDF

X_train = items.iloc[:, -19:].values
X_train.shape
```

```python
from sklearn.feature_extraction.text import  TfidfTransformer
transformer = TfidfTransformer(smooth_idf=True, norm ='l2')
tfidf = transformer.fit_transform(X_train.tolist()).toarray()
```

```python
tfidf
```

```python
tfidf.shape
```

```python
# for each user, find out which movies and how much they rated them. 

def get_items_rated_by_user(rate_matrix, user_id):
    """
    in each line of rate_matrix, we have info: user_id, item_id, rating (scores), time_stamp
    we care about the first three values: user_id, item_id, rating
    return (item_ids, scores) rated by user user_id
    """
    y = rate_matrix[:,0] # all users
    # item indices rated by user_id
    # we need to +1 to user_id since in the rate_matrix, id starts from 1 
    # while index in python starts from 0
    ids = np.where(y == user_id +1)[0] 
    item_ids = rate_matrix[ids, 1] - 1 # index starts from 0 
    scores = rate_matrix[ids, 2]
    return (item_ids, scores)
```

```python
# rating of each item is predicted by:

Yhat = tfidf.dot(W) + b 
```

4. Evaluate Model

```python
def evaluate(Yhat, rates, W, b):
    ''' 
    Use the Root Mean Squared Error (RMSE), 
    which is the square root of the mean squared error. 
    The error is calculated as the difference of true rating and predicted rating
    '''
    se = 0 #squared error
    cnt = 0 #count variable
    for n in range(users.shape[0]):
        item_ids, scores_truth = get_items_rated_by_user(rates, n) #ground truth
        scores_pred = Yhat[item_ids, n] #predicted rating
        e = scores_truth - scores_pred #error
        se += (e*e).sum(axis=0)
        cnt += e.size
    return np.sqrt(se/cnt)
```

```python
print ('RMSE for training:', evaluate(Yhat, rate_train, W, b))
print ('RMSE for test    :', evaluate(Yhat, rate_test, W, b))
```
